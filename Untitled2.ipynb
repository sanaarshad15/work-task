```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
data = pd.read_csv('/content/Automobile_data (1).csv')
```

```python
# Show information about data
print(data.info())
```

```python
print(data.describe())
```

```python
# Find duplicates
duplicates = data[data.duplicated()]
```

```python
duplicates
```

```python
# Drop duplicates
data = data.drop_duplicates()
```

```python
data
```

```python
# Replace '?' with NaN values
data = data.replace('?', pd.NA)
```

```python
data
```

```python
# Find missing values
print(data.isnull().sum())
```

```python
data['normalized-losses'] = data['normalized-losses'].fillna(data['normalized-losses'].mean())
```

```python
data['normalized-losses']
```

```python
# Drop null values from 'price' column
data = data.dropna(subset=['price'])
```

```python
data
```

```python
# Plot 'normalized-losses' aur 'symboling' columns
plt.scatter(data['normalized-losses'], data['symboling'])
```

```python
plt.show()
```

```python
data['normalized-losses'] = pd.to_numeric(data['normalized-losses'], errors='coerce')
```

```python
data
```

```python
# Drop outliers
data = data[(data['normalized-losses'] < 100) & (data['symboling'] < 10)]
```
